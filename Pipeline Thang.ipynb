```python
# Used to select cat_features or num_features to fit into preprocessing pipeline of their type
class FeatureSelector( BaseEstimator, TransformerMixin ): 
    def __init__( self, feature_names ):
        self._feature_names = feature_names 
    
    def fit( self, X, y = None ):
        return self 
    
    def transform( self, X, y = None ):
        return X[ self._feature_names ] 
```

```python
class NumTransformer( BaseEstimator, TransformerMixin):
    def __init__(self):
        return None
    #def to_nan(self,obj):
        #return np.NaN
    def fit( self, X, y = None ):
        return self 
    def transform(self,X,y=None):
        X.replace('None',np.nan)
        X.replace('na',np.NaN)
        X['FIELD_3'].replace(-1.0, np.NaN, inplace=True)
        #why only field3 ???
        return X.values
                
        #X=X.replace('Tỉnh Vĩnh phúc', 'Tỉnh Vĩnh Phúc', inplace=True)
        
```

```python
cat_features = []
num_features = []
for col in cs:
    if cs[col].dtype in ['int64','float64']:
        num_features.append(col)
    else: 
        cat_features.append(col)
num_features.remove('id')
num_features.remove('label')
```

```python
import numpy as np 
import pandas as pd


cs2=pd.read_csv('D:/Kalapa competition/train.csv')
train=cs2.drop(['label'],axis=1)
test=pd.read_csv('D:/Kalapa competition/test.csv')
#Join train file and test file into one BIG dataframe
frames = [train,test]
trainTest = pd.concat(frames)
trainTest.head()


df=pd.DataFrame(CatTransformer().transform(trainTest[cat_features]))
df.columns=cat_features

from sklearn import preprocessing
le = preprocessing.LabelEncoder()

# Create a lookupTable from the BIG dataframe
lookupTable = {}
for col in df:
    curCol=df[col].unique().tolist()
    if np.nan in curCol:
        curCol.remove(np.nan)
    curEncoded=le.fit_transform(curCol).tolist()
    curDict=dict(zip(curCol,curEncoded))
    d={str(col):curDict}
    lookupTable.update(d)
```

```python
lookupTable
```

```python
#Create a Custom Transformer to encode categorical features based on LookupTable
class CustomEncoder(BaseEstimator, TransformerMixin):
    def __init__(self):
        return None
    
    def fit( self, X, y = None ):
        return self 
    
    def transform(self, X, y = None):
        temp_df=pd.DataFrame(X)
        temp_df.columns=cat_features
        for col in temp_df:
            hashMap = lookupTable[col]
            newContents = []
            for item in temp_df[col]:
                newContents.append(hashMap.get(item))
            temp_df[col]=newContents
        
        return temp_df.values 
            
```

```python
#Train Split Test
from sklearn.model_selection import train_test_split
X = cs.drop(['label','id'],axis=1)
y = cs['label']
X_train, X_test, y_train, y_test = train_test_split(
X, y, test_size=0.20, random_state=42)
```

```python
#Put everything into pipelines
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OrdinalEncoder
from sklearn import preprocessing

from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn import svm
from sklearn.tree import DecisionTreeClassifier



cat_transformer= Pipeline(steps=[
    ('cat_selector',FeatureSelector(cat_features)),
    ('cat_transfomer',CatTransformer()),
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('custom', CustomEncoder()),
])

num_transformer = Pipeline(steps=[
    ('num_selector',FeatureSelector(num_features)),
    ('num_transfomer',NumTransformer()),
    ('imputer', SimpleImputer(strategy='median')),
    ('normalizer', preprocessing.StandardScaler())])

#     preprocessor = ColumnTransformer(
#         transformers=[
#             ('cat', cat_transformer, cat_features),
#             ('num', num_transformer, num_features)])

preprocessor = FeatureUnion(transformer_list = [ ( 'categorical_pipeline', cat_transformer ), 
                                                  
                                                  ( 'numerical_pipeline', num_transformer ) ] )


pipe_knn = Pipeline([('pre', preprocessor),
                    ('clf', KNeighborsClassifier())])
```

```python
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score
```

```python
pipe_knn.fit(X_train,y_train)
y_pred=pipe_knn.predict(X_test)
    #print(str(algo))
print('Accuracy Score : ' + str(accuracy_score(y_test,y_pred)))

print('Precision Score : ' + str(precision_score(y_test,y_pred)))

print('Recall Score : ' + str(recall_score(y_test,y_pred)))

print('F1 Score : ' + str(f1_score(y_test,y_pred)))
```
